```python
def plot_quartos(bairro):
    plt.plot(range(len(bairro.qt1())),bairro.qt1(),'-b',bairro.qt2(),'-g',bairro.qt3(),'-r')
    plt.title(bairro.nome)
    plt.xlabel('meses desde 2008...')
    plt.ylabel(r'$mil R\$/m^2$')
    plt.legend(['1 quarto','2 quartos','3 quartos '],loc='upper left')

def plot_bairro(bairro,num_quartos=0):
    if type(bairro) == str:
        bairro = db.Bairro('rio de janeiro','rio de janeiro', bairro)
    plt.plot(range(len(bairro.mes)), bairro.qt(num_quartos))
```

```python
rj = db.Cidade('rio de janeiro','rio de janeiro')
plot_quartos(rj.bairro('tijuca'))
```

```python
plot_quartos(rj.bairro('centro'))
```
