```python
# Variáveis sócio econômicas, por bairro.
df_soc = z.get('select *  from _var_socioecon ',None)            
```

```python
# Unir imóveis com variáveis sócio econômicas por bairro.
df = z.pd.merge(df,df_soc, left_on='id_bairro_g', right_on='gid', how='inner')
```

```python
# Variáveis espaciais.
df_dist = z.get('select id, dist_favela,dist_bombeiro,dist_centro,' + \
    'dist_centro_lng,dist_centro_lat,dist_delegacia,dist_lagoa,' + \
    'dist_logradouro,dist_metro,dist_praia,dist_saude_publica,' + \
    'dist_saude_privada,dist_trem from vw_distancia', None)
```

```python
# Variáveis intrínsecas.
df_int = z.get('select * from _var_intrinseca', None)
cols = df_int.columns.tolist()
cols_ = ['dm_'+i.replace(' ','_') for i in cols if i <> 'id']
df_int.columns = ['id'] + cols_
```

```python
# Unir com variáveis espaciais e intrínsecas.
df_int.index = df_int.id
df_dist.index = df_dist.id
df.index = df.id
del df['id'], df_dist['id'], df_int['id']
df = z.pd.concat([df,df_dist,df_int], axis=1, join='inner')
```

```python
df.head()
```

```python
len(df)
```

```python
formula = 'df.preco ~  C(df.bairro_g) '
lm = sm.formula.ols(formula, df).fit()

print lm.summary()
```

```python
# Lista de Bairros sem significância.
bp = lm.pvalues.copy()
bp.index = [i.replace('C(df.bairro_g)[T.', '').replace(']','')  for i in bp.index.tolist()]
c = bp > 0.05
bpi = bp[c]
bpi
```

```python
s = g.size()
x = s.index.isin(bpi.index.tolist())
df2 = z.pd.DataFrame({'pvalue':bpi, 'qtd':s[x]}, index=bpi.index)
scatter(df2.qtd, df2.pvalue)
```

```python
lmb = lm.params + lm.params.Intercept
lmb.drop('Intercept', inplace=True)
gb = df.groupby('bairro_g')
mean, median = gb.preco.mean() , gb.preco.median()
```

```python
df1 = z.pd.DataFrame({'media':mean, 'mediana':median})
df1['b'] = df1.index
df2 = z.pd.DataFrame({'coef':lmb})
df2.index = [i.replace('C(df.bairro_g)[T.', '').replace(']','')  for i in df2.index.tolist()]
df2['b'] = df2.index
dfb = z.pd.merge(df1,df2, how='outer', on='b')
dfb.index = dfb.b
del dfb['b']
dfb
```

```python
idx_bairro = lm.params.index.map(lambda x: x.find('df.bairro_g') <> -1) 
s = lm.params[idx_bairro]
s.sort(ascending=False)
s
```
