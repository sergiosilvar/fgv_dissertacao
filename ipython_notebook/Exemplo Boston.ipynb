```python
x = boston.data[:,5]
x = np.array([[v] for v in x])
y = boston.target
slope,error,_,_ = np.linalg.lstsq(x,y)
xp = [min(x), max(x)]
yp = slope*xp
scatter(x,y,color='r')
plot(xp,yp,'b')
errors = [error]
```

```python
x = boston.data[:,5]
x = np.array([[v,1] for v in x]) # we now use [v,1] instead of [v]
y = boston.target
(slope,bias),error,_,_ = np.linalg.lstsq(x,y)
xp = [x[:,0].min(), x[:,0].max()]
yp = dot(slope,xp)+bias
scatter(x[:,0],y,color='r')
plot(xp,yp,'b')
errors.append(error)
```

```python
x = boston.data
# we still add a bias term, but now we must use np.concatenate,which
# concatenates two arrays/lists because we
# have several input variables in v
x = hstack((boston.data,ones((shape(boston.data)[0],1))))#array([concatenate(v,[1]) for v in boston.data])
y = boston.target
s,error,_,_ = np.linalg.lstsq(x,y)
errors.append(error)
```

```python
plot(errors)
```

## Usando Scikit Learn

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(fit_intercept=True)
lr.fit(x,y)
p = lr.predict(x)
e = p-y
rmse = sqrt(mean(e**2))
rmse
```

```python
from sklearn.cross_validation import KFold
kf = KFold(len(x), n_folds=10)
err = 0
for train,test in kf:
    lr.fit(x[train],y[train])
    p = lr.predict(x[test])
    e = p-y[test]
    err += np.sum(e*e)
rmse_10cv = np.sqrt(err/len(x))
print('RMSE on 10-fold CV: {}'.format(rmse_10cv))
```
