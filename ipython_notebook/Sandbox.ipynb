```python
def plotfit(degree):
    p = polyfit(x,y2,degree)
    yh = polyval(p, x)
    scatter(x,y2,color='r')
    plot(x,yh)
    rmse = sqrt(sum((yh-y)**2)/len(y))
    text(4,1,'RMSE: {:.4f}'.format(rmse))
```
