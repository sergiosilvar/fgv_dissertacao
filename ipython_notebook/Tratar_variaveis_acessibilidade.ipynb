```python
%pylab inline

import zap_util as z
z.set_style()
DIR = './../texto/img/'
```

```python
df = z.get('select * from vw_distancia inner join imovel_sel using(id)')
```

```python
dfd = df[[i for i in df.columns.tolist() if i.find('id_') ==-1]]
dfd.describe()
```

```python
def plot(column, legend):
    z.plot_boxhist(dfd[column], u'Variável '+column, 
        u'Distância {} em Km.'.format(legend)  )
    savefig(DIR+'var_'+column+'_boxhist.png')
```
