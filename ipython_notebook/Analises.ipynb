```python
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing

def linear_reg(X,y):
    '''
        Helper para usar regressao linear do sklearn.
    '''
    lr = LinearRegression(fit_intercept=True)
    lr.fit(X,y)
    p = lr.predict(X)
    rmse = sum((p-y)**2)/len(p)
    return rmse

errors = []
rmse = linear_reg(df[['lat','lng']],df.preco)
errors.append(rmse)
```

```python
e = linear_reg(df[['area','quartos','suites']], df.preco)
errors.append(e)
```

```python
e = linear_reg(df[['area','quartos','suites','condominio']], df.preco)
errors.append(e)
```

```python
e = linear_reg(df[['area', 'quartos', 'suites', 'condominio', 'garagem' ]], df.preco)
errors.append(e)
```

```python
e = linear_reg(df[['area', 'quartos', 'suites', 'condominio', 'garagem', 'lat', 'lng']], df.preco)
errors.append(e)
```

```python
def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d'%int(height),
                ha='center', va='bottom')
fig,ax = subplots()        
rct = ax.bar(range(len(errors)),errors, width=0.3)
autolabel(rct)
```
