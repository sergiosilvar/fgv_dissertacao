```python
# Remover os que não tem condomínio e quartos.
df = df[~df.condominio.isnull()]
df = df[~df.quartos.isnull()]

# Remover os que tem mais de 5 quartos.
df = df[df.quartos <=4]

# Remover área inferior a 20.
df = df[df.area >= 20]

df = df[df.m2 < 6* df.m2.std()]
df = df[df.preco < 6* df.preco.std()]

# Tratar entradas nulas de garagem e suites como valor ZERO.
df.garagem.fillna(value=0, inplace=True)
df.suites.fillna(value=0, inplace=True)

# Tratar entradas nulas de rua e cep como "ND em <Bairro>"
df.cep.loc[df.rua.isnull()] = 'ND em ' + df[df.rua.isnull()].bairro_g
df.cep.loc[df.cep.isnull()] = 'ND em ' + df[df.cep.isnull()].bairro_g

# Remover os que tem mais vagas de garagem do que quartos.
df = df[df.garagem <= df.quartos]

# Verificar se todos os NaN foram resolvidos.
df.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing

def linear_reg(X,y):
    '''
        Helper para usar regressao linear do sklearn.
    '''
    lr = LinearRegression(fit_intercept=True)
    lr.fit(X,y)
    p = lr.predict(X)
    rmse = sum((p-y)**2)/len(p)
    return rmse

errors = []
rmse = linear_reg(df[['lat','lng']],df.preco)
errors.append(rmse)
```

```python
e = linear_reg(df[['area','quartos','suites']], df.preco)
errors.append(e)
```
