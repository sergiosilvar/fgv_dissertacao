```python
%pylab inline

import zap_util as z
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Lista de modelos
rmse = {}
```

```python
df = z.get('select * from vw_imovel')
numeric_columns = (df.dtypes[(df.dtypes==float64) | (df.dtypes==int64)]).index
df = df[numeric_columns]
df.columns
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
df.isnull().sum()
```

# Regressão apenas com latitude e longitude.

```python

formula = 'log(df.preco) ~ df.lat+ df.lng'
lm = sm.formula.ols(formula, df).fit()
rmse['lat e lng']= z.rmse(lm.resid)
print lm.summary()

```

```python
z.plot_residual(lm)
```

# Regressão com dados básicos dos anúncios.

```python

formula = 'log(df.preco) ~ df.area + df.condominio + df.garagem + df.quartos + df.suites+ df.lat+ df.lng'
lm = sm.formula.ols(formula, df).fit()
rmse['basico']= z.rmse(lm.resid)
print lm.summary()
```

```python
z.plot_residual(lm)
```

# Regressão com variáveis intrínsecas binárias.

```python
dc = z.get('select * from _carac_explo')

# Adicionar prefixo "dm" para indicar variável dummy e remover espaços dos nomes.
dc.columns = ['dm_'+c.replace(' ','_') for c in dc.columns]

# Verificar validade das variáveis dummies. O valor máximo por coluna deve ser igual a 1.
m = dc.max(axis=0) 
m = m[m > 1].index.tolist()
col_excluded =  set(m)
for c in col_excluded:
    del dc[c]
m = dc.max(axis=0) 
m = m[m > 1].index.tolist()
m
```

```python
d = z.pd.concat([df,dc],1,'inner')
```

```python
# Aplicar o modelo.
lm = sm.formula.ols(z.prep_formula(d,'d', func='log')[0], d).fit()
rmse['intrinseca']= z.rmse(lm.resid)
print lm.summary()


```

```python
z.plot_residual(lm)
```

# Encontrar características autocorrelacionadas.

```python
# Caractarísticas dummies não são adequadas par correlações.
d = z.pd.concat([df,dd,dv],1,'inner')
# Variávies categóricas numéricas.
col_categoricas = ['garagem','quartos','suites']

df_dum = []

for c in col_categoricas:
    dum = z.pd.get_dummies(d[c])
    dum.columns = ['dm_'+c+'_'+str(int(i)) for i in dum.columns]
    df_dum.append(dum)

    
# Selecionar colunas que não são identificadores e variáveis dependentes.
cols_excluded = set([c for c in d.columns if c.find('id_')==0])
cols_excluded = cols_excluded.union(set(['preco','m2']+col_categoricas))
    
d = z.pd.concat([df,dc,dd,db]+df_dum,1,'inner')


z.print_autocorr(d,cols_excluded)





```

# Repetir modelo completo excluindo as colunas auto correlacionadas.

```python
# Atualizar coluna a ser excluída do modelo.
valid_cols = list(set(d.columns.tolist())-cols_excluded)

z.print_autocorr(d[valid_cols])
```

```python
# Executar o modelo.
d2 = d[valid_cols+['preco']]
lm = sm.formula.ols(z.prep_formula(d2,'d2', func='log')[0], d2).fit()
rmse['completo, sem autocorr ']= z.rmse(lm.resid)
print lm.summary()

```

```python
z.plot_residual(lm)
```

Vemos que os muitos p-value > 0.05, constando que $H_0 : mean(X_i) = 0$ não pode ser refutada.

```python
# Características cujo p-value é inferior a 0.05.
cols_pvalue = lm.pvalues[lm.pvalues <= 0.05].index.tolist()
try:
    cols_pvalue.remove('Intercept')
except:
    pass

# Retirar prefixo colocado pelo statsmodels.
cols_pvalue = [c.replace('d2.','') for c in cols_pvalue  ]

# Executar o modelo.
d2 = d2[cols_pvalue+['preco']]
lm = sm.formula.ols(z.prep_formula(d2,'d2', func='log')[0], d2).fit()
rmse['completo, sem autocorr e pvalue']= z.rmse(lm.resid)
print lm.summary()
```
